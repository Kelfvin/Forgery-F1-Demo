```python
import cv2
import matplotlib.pyplot as plt

gt_path = "./asset/GT.png"

img = cv2.imread(gt_path, cv2.IMREAD_GRAYSCALE)

fig, ax = plt.subplots(1, 1, figsize=(10, 10))

ax.imshow(img, cmap='gray')
plt.show()
```

```python
# 图像转换成 RGB 格式
img_rgb = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)


def split_image(img, i):
    h,w = img.shape[0], img.shape[1]
    cell_h, cell_w = h//i, w//i

    # 绘制红线
    for i in range(1, i):
        img[i*cell_h, :] = [255, 0, 0]
        img[:, i*cell_w] = [255, 0, 0]

    return img


img_split = split_image(img_rgb, 4)

fig, ax = plt.subplots(1, 1, figsize=(10, 10))

ax.imshow(img_split)
plt.show()
```
